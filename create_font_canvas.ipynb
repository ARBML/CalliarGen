```python
import os
from glob import glob 
fonts = glob("fonts/**.ttf")
font_names = open("fonts/font_names.txt", 'r').read().splitlines()
print(font_names)
```

```python
import json
with open('words_freq.txt', 'r') as f:
    data = json.load(f)
print(len(data))
```

```python
words = list(data.keys())
freqs = list(data.values())
total = sum(freqs)
freqs = [freq / total for freq in freqs]
```

```python
from numpy.random import choice
def sample():
    return choice(words, p = freqs)
```

```python
from PIL import Image
from PIL import ImageDraw
import matplotlib.pyplot as plt
from PIL import ImageFont
from IPython import display
import random
import numpy as np
# Open an Image
font_size = random.randint(50, 100)
bg = np.random.choice(glob('fonts-bgs/**.jpg'))
img = Image.open(bg)
W, H = img.size
fontColor = "black"
font_type = np.random.choice(fonts)
font_name = font_names[int(font_type[6:-4]) - 1]
print(font_name)
text = sample()
font = ImageFont.truetype(f"{font_type}",font_size)
# Call draw Method to add 2D graphics in an image
draw = ImageDraw.Draw(img)
w, h = draw.textsize(text, font=font)
draw.text(((W-w)/2, (H-h)/2), text, font=font, fill=fontColor) 
img.save('tmp.png')
display.Image('tmp.png')
```
