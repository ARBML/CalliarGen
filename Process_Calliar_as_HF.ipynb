```python
from glob import glob 
import matplotlib.pyplot as plt
from PIL import Image
import shutil



path_to_images = "full_data/images/*.jpg"

train_images = []
train_texts = []

test_images = []
test_texts = []

train_size = 2200

for i, image_path in enumerate(glob(path_to_images)):
  image_name = image_path.split('/')[-1]
  text = image_name[:-4]
  image_id = f"{i:05d}.jpg"
  if '_' in text:
    text = text.split('_')[-1]
  if i < train_size:
    shutil.copyfile(image_path, f"data/train/{image_id}")
    train_images.append(image_id)
    train_texts.append(text)
  else:
    shutil.copyfile(image_path, f"data/test/{image_id}")
    test_images.append(image_id)
    test_texts.append(text)
```

```python
import pandas as pd
train_df = pd.DataFrame({'file_name':train_images, 'caption':train_texts})
test_df = pd.DataFrame({'file_name':test_images, 'caption':test_texts})

train_df.to_csv('data/train/metadata.csv', index = False)
test_df.to_csv('data/test/metadata.csv', index = False)
```
